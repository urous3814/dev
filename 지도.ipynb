```python
import pandas as pd
import numpy as np
import folium
import json
```

```python
rk = pd.read_csv('roadkill.csv', encoding='euc-kr')
```

```python
rk
```

```python
rk = rk.drop(['본부명','지사명', '구간', '노선명', '방향', '5km'], axis=1)
rk = rk.dropna(axis=0)
```

```python
rkmap=folium.Map(location=[35.95, 123.25], zoom_start=10)
```

```python
for i in range(len(rk)):
  la = rk['위도'][i]
  lo = rk['경도'][i]
  name = str(rk['발생건수'][i])
  m=folium.Marker([la, lo], popup=name)
  m.add_to(rkmap)
```
